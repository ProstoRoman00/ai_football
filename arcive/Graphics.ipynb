```python
import matplotlib.pyplot as plt
import numpy as np

import csv
nc=[]*6
for i in range(0,7):
    nc.append([])
with open('results_days.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    next(csv_reader)
    for row in csv_reader:
        #names = ['AI1 score', 'AI2 score', 'AI3 score','AI1 result', 'AI2 result', 'AI3 result']
        for i in range(0,6):
            nc[i].append(int(row[i]))
        nc[6].append(row[6])    
for i in range(0,3):
    fig, ax = plt.subplots(figsize=(8, 4))
    ax.stackplot(nc[6], nc[i+3], labels=['Complete score'])
    ax.stackplot(nc[6], nc[i], labels=['Game result'])
    ax.set_title('Neural Network #'+str(i+1)+' Result')
    ax.legend(loc='upper left')
    ax.set_ylabel('Total predicted')
    ax.set_ylim(ymin=0, ymax=20)
    fig.tight_layout()
```
