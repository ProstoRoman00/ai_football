```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Reshape, Dropout, Dense, Flatten
from tensorflow.keras.models import Sequential, Model, load_model
import numpy as np
import os 
import csv
```

```python
model = load_model('model.h5')
model1 = load_model('model1.h5')
model2 = load_model('model2.h5')
```

```python
results=[]
f2 = open("../results.csv")
res=f2.readlines()[1:]
with open('../games.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row,teams in zip(csv_reader,res):
        row=[list(map(float, row))]
        teams=teams.split(",")[0:3]
        pre = model.predict(row)
        pre1 = model1.predict(row)
        pre2 = model2.predict(row)
        results.append([teams[0],teams[1],teams[2].replace('\n',''),pre[0][0],pre[0][1],pre1[0][0],pre1[0][1],pre2[0][0],pre2[0][1]])
```

```python
with open('../results.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["t1","VS","t2","m1t1","m1t2","m2t1","m2t2","m3t1","m3t2","a1","a2"])
    for game in results:
        writer.writerow(game)
```
